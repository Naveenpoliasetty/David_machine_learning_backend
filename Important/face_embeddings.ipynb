```python
from deepface import DeepFace
import numpy as np
import cv2
```

```python
embedding_objs = DeepFace.represent(
  img_path = "/Users/naveenpoliasetty/Downloads/David2.0/david2.0/Frames/frame_1.jpg"
)
face_embeddings = [embedding_objs[x]['embedding'] for x in range(len(embedding_objs))]
```

```python
embds = DeepFace.represent('/Users/naveenpoliasetty/Downloads/David2.0/david2.0/faces/213J1A42C9.jpg')
embeddings = [embds[x]['embedding'] for x in range(len(embds))]
```

```python
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)

    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    
    if norm_vec1 == 0 or norm_vec2 == 0:
        return 0.0
    cosine_sim = dot_product / (norm_vec1 * norm_vec2)
    
    return cosine_sim

def compare_faces(embedd1, indiviual_faces_embeddings):
    scores, dot_scores = [], []
    for x in embedd1:
        scores.append(cosine_similarity(indiviual_faces_embeddings, x))
        dot_scores.append(np.dot(indiviual_faces_embeddings, x))


    for x, y in zip(scores, dot_scores):
        if x > 0.5:
            print('True cosine similarity')
     
        if y > 0.5:
            print('True dot product')
    return np.argmax(scores), np.argmax(dot_scores), scores, dot_scores
```

```python
compare_faces(face_embeddings, embeddings)
```
